```python
# importando as bibliotecas
import pandas as pd
import os
import numpy as np
```

```python
#Recuperando todos os arquivos de excel presentes na pasta 
data_df = None
nomePlanilha =[]
for filename in os.listdir('/home/bia/Teste_Python/'):
   if filename.endswith(".xlsx"):
      nomePlanilha.append(filename)
      data_df = pd.concat([ pd.Series( pd.read_excel(f'/home/bia/Teste_Python/{filename}', sheet_name=None  ) ) , pd.Series( data_df) ] )
```

```python
df_final = None
tam = len(data_df) 
if(tam % 2 != 0):
    tam = tam - 1
    data_df[tam]['Data Período'] = nomePlanilha[tam].replace('_', '/').replace('Teste', '01').replace('.xlsx', "")
    df_final = data_df[tam]
    tam = tam - 1
while(tam >= 0):
    data_df[tam]['Data Período'] = nomePlanilha[tam].replace('_', '/').replace('Teste', '01').replace('.xlsx', "")
    df_final = pd.concat([data_df[tam], df_final] ).reset_index()
    tam -= 1
    df_final = df_final.drop("index", axis=1)
```

```python
# Ordenando para que a data mais recente fique visivel primeiro
df_final = df_final.sort_values(by='Data Período' , ascending=False)
```

```python
#Renomeando colunas
df_final.columns = ['CC', 'Licença', 'Contrato', 'Login', 'Data Período']
```

```python
# Descobrindo o tipo dem que a coluna de centro de custo foi salva como
type(df_final['CC'][0])

#transformando int em str
df_final['CC'] = df_final['CC'].astype(str)
type(df_final['CC'][2])
```

```python
#preenchendo os zeros a esquerda
for x in range(len(df_final)):
    while(len(df_final.CC[x]) < 4):
            df_final.CC[x] =  df_final.CC[x].replace( df_final.CC[x], f'0{df_final.CC[x]}')
```

```python
# Reajustando o index
df_final = df_final.reset_index().drop("index", axis=1)
```

```python
# Padronizando a coluna de Data
df_final['Data Período'] = pd.to_datetime(df_final['Data Período'] , format='%d/%m/%Y') 
```

```python
# Definindo a coluna Periodo (Mês por extenso)
df_final['Período'] = None
meses = ['JANEIRO', 'FEVEREIRO', 'MARÇO', 'ABRIL' , 'MAIO' , 'JUNHO', 'JULHO' , 'AGOSTO', 'SETEMBRO' , 'OUTUBRO', 'NOVEMBRO', 'DEZEMBRO']
for x in range(0, len(df_final)):
    mes = int( df_final['Data Período'][x].strftime("%-m") ) - 1
    df_final['Período'][x] = meses[mes]
```

```python
dt_inteiro = pd.merge(df_final, df_solicitantes, how='inner')
```
